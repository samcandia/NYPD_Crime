```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Susp_Age_Group.csv", encoding='utf-8', low_memory=False)
```

```python
top_5_felony = df[df['LAW_CAT_CD'] == 'FELONY']['OFNS_DESC'].value_counts().head(5)
top_5_misdemeanor = df[df['LAW_CAT_CD'] == 'MISDEMEANOR']['OFNS_DESC'].value_counts().head(5)
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.barplot(x=top_5_felony.values, y=top_5_felony.index, palette="viridis")
plt.xlabel("Count")
plt.ylabel("Felonies")
plt.title("Top 5 Felonies in 2023")
plt.show()
```
